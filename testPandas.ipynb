```python
def convert_to_datetime(df, columns):
  """
  Converts a list of columns in a DataFrame to datetime format.

  Args:
      df: The DataFrame to convert.
      columns: A list of column names to convert.

  Returns:
      A new DataFrame with the specified columns converted to datetime.
  """
  try:
    # Attempt to convert using a common date format
    return df[columns].apply(pd.to_datetime, format='%Y-%m-%d')
  except ValueError:
    # If the common format fails, try a more flexible format
    return df[columns].apply(pd.to_datetime, errors='coerce')

```

```python
def rename_columns(df, old_names, new_names):
  """
  Renames a list of columns in a DataFrame based on a mapping between old and new names.

  Args:
      df: The DataFrame to modify.
      old_names: A list of column names to be renamed.
      new_names: A list of new names for the corresponding columns.

  Raises:
      ValueError: If the lengths of `old_names` and `new_names` are not equal.
  """

  if len(old_names) != len(new_names):
    raise ValueError("Number of old and new names must be equal.")

  # Create a dictionary for renaming
  rename_dict = dict(zip(old_names, new_names))

  # Rename columns using the dictionary
  df.rename(columns=rename_dict, inplace=True)
```
